# Sales Brochure Generator with OpenAI Chat Completions API

This notebook builds a simple AI-powered sales brochure generator.

The goal is to learn how to:
- collect structured business inputs
- design a strong prompt
- generate marketing copy
- format the output as a brochure

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display

load_dotenv()

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
MODEL = "gpt-4o-mini"
```

```python
# Define company information

company_info = {
"company_name": "YugData Solutions",
"industry": "Healthcare Data Engineering",
"target_audience": "Healthcare organizations, PBMs, and analytics teams",
"main_product": "a cloud-based healthcare data platform",
"pain_points": [
    "manual reporting",
    "slow claims data processing",
    "poor data quality",
    "difficulty integrating pharmacy benefit data"
],
"benefits": [
    "automated ETL pipelines",
    "faster analytics",
    "improved data quality",
    "secure cloud-based processing"
],
"tone": "professional, clear, and persuasive",

}
```

```python
def build_brochure_prompt(company: dict) -> str:
    return f"""
Create a professional sales brochure for the following company.

Company Name:
{company["company_name"]}

Industry:
{company["industry"]}

Target Audience:
{company["target_audience"]}

Main Product:
{company["main_product"]}

Customer Pain Points:
{", ".join(company["pain_points"])}

Product Benefits:
{", ".join(company["benefits"])}

Tone:
{company["tone"]}

Brochure Requirements:
- Create a strong headline
- Write a short opening paragraph
- Include 4 key benefits
- Include a section explaining why customers should choose this company
- Include a short call to action
- Format the brochure in clean Markdown
"""
```

```python
# Generate brochure
prompt = build_brochure_prompt(company_info)

response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {
            "role": "system",
            "content": "You are an expert B2B marketing copywriter."
        },
        {
            "role": "user",
            "content": prompt
        }
    ],
    temperature=0.7,
    
)
brochure = response.choices[0].message.content
display(Markdown(brochure))
```

```python
# show token usage
print(response.usage)
```

```python
# Improve with stricter structure
def build_structured_brochure_prompt(company: dict) -> str:
    return f"""
    You are creating a sales brochure for a B2B software company.

    Use this exact structure:

    # [Compelling Brochure Title]

    ## Who We Help
    Explain the target audience.

    ## The Problem
    Describe the pain points clearly.

    ## Our Solution
    Explain the product.

    ## Key Benefits
    Use exactly 4 bullet points.

    ## Why Choose {company["company_name"]}
    Explain differentiation.

    ## Call to Action
    End with a short persuasive CTA.

    Company details:
    - Company Name: {company["company_name"]}
    - Industry: {company["industry"]}
    - Target Audience: {company["target_audience"]}
    - Product: {company["main_product"]}
    - Pain Points: {", ".join(company["pain_points"])}
    - Benefits: {", ".join(company["benefits"])}
    - Tone: {company["tone"]}

    """
    
```

```python
structured_prompt = build_structured_brochure_prompt(company_info)

response = client.chat.completions.create(
    model=MODEL,
    messages=[
        {"role": "system", "content": "You are a senior B2B SaaS marketing strategist."},
        {"role": "user", "content": structured_prompt}
    ],
    temperature=0.5,
)

structured_brochure = response.choices[0].message.content
display(Markdown(structured_brochure))
        
```

```python
# Save brochure to markdown file
output_path = "../sales_brochure.md"

with open(output_path, "w", encoding="utf-8") as file:
    file.write(structured_brochure)

print(f"Brochure saved to {output_path}")

```

## What I Learned

- Chat Completions can generate structured business content.
- Better prompts create more predictable output.
- Lower temperature gives more controlled responses.
- Markdown formatting makes generated content easier to reuse.
- Structured inputs help reduce vague or generic responses.
